**Example 1: Multiple Linear Regression in R**

The following example gives an R implementation of Multiple Linear Regression in R (Data for independent and Dependent variables are generated since I don't have data available) and how to interpret the linear model summary table. And also how to reduce the number of variables based from the $\alpha$ level of significance set.

```r
#Multiple Linear Regression
#Generating data to be fitted
set.seed(45)
n <- 100

#Independent Variables
Result_df <- data.frame(
  Var_0 = rnorm(n, mean = 75, sd = 10),
  Var_1  = rnorm(n, mean = 120, sd = 25),
  Var_2  = sample(2:5, n, replace = TRUE),
  Var_3  = sample(1:3, n, replace = TRUE),
  Var_4  = rnorm(n, mean = 20, sd = 10),
  Var_5  = sample(0:2, n, replace = TRUE),
  Var_6  = rnorm(n, mean = 8, sd = 3),
  Var_7  = rnorm(n, mean = 50, sd = 15)
)

#Creating the independent variable
Result_df$Y <- 4.8 +
  0.30 * Result_df$Var_0 +
  0.05 * Result_df$Var_1 -
  1.20 * Result_df$Var_2 +
  2.00 * Result_df$Var_3 +
  0.40 * Result_df$Var_4 -
  3.00 * Result_df$Var_5 +
  1.50 * Result_df$Var_6 +
  0.10 * Result_df$Var_7 +
  rnorm(n, mean = 0, sd = 5)    #adding noise

#Fitting to the linear model
model <- lm(
  Y ~ Var_0 + Var_1 + Var_2 + Var_3 + Var_4 + Var_5 + Var_6 + Var_7,
  data = Result_df
)

summary(model)
```

Observe that our adjusted $R^2$ is less than $R^2$. Adjusted $R^2$ is smaller since it includes a penalty for adding independent variables to our model
<br><br>
Let's say that our $\alpha$ level of significance is 0.05, then observe that the variables having very low p - values (less than our $\alpha$ level)  are Var_0, Var_1, ..., Var_6. Only Var_7 did not made it to the cut.
<br><br>
This means that the variables  Var_0, Var_1, ..., Var_6 are contributing a significant impact to our model. Only Var_7 did not made a significant impact. In other words, independent variables having very low p- values (lower than our $\alpha$ level of significance) are doing a significant impact to the model
<br><br>
When we redo the regression, we have the following

**Example 2: LOESS Model Fitting and tuning for different span parameter**

In this example, we generate our data (I don't have a raw data so I generate it just for the sake of example in implementing LOESS).

Next, we fit our data to follow LOESS model fit

```r
# Generate sample data
a <- 0
b <- 6
h <- 0.05

x <- seq(a, b, h)
n <- length(x)

# Pre-allocate blank vector y
y <- numeric(n)

for (i in 1:n) {
  y[i] <- 0.85* x[i]^3 - 2 * x[i]^2 + rnorm(1, mean = 0, sd = 10.5)
}

plot(x, y, pch = 1)
```

Fitting the data to LOESS model

```r
#fitting to loess model

fit_loess <- loess(y ~ x)
```

```r
#Generating data for plotting
x_forpred <- seq(min(x), max(x), length = 150)
y_pred <- predict(fit_loess, newdata = data.frame(x = x_forpred))

#Plotting
plot(x, y, pch = 19, col = "gray")
lines(x_forpred, y_pred, col = rgb(1,0,0,0.5), lwd = 3)
```

Generating different LOESS curve for different span

```r
#Observing loess fit for different span parameter

plot(x, y, pch = 19, col = "gray")

spans <- c(0.5, 0.75, 3)
colors <- c("pink", "light blue", "light green")


#Iterating for different spans and plotting
for (i in seq_along(spans)) {
  fit <- loess(y ~ x, span = spans[i])
  lines(x_forpred, predict(fit, data.frame(x = x_forpred)),
        col = colors[i], lwd = 2)
}

legend("topleft",
       legend = paste("span =", spans),
       col = colors, lwd = 2)
```

**Example 3: Kernel Regression (kernel = Gaussian) and tuning for bandwidth**

In this example, we will use the same x and y values used in LOESS model (Example 2)

```r
# Generate sample data
a <- 0
b <- 6
h <- 0.05

x <- seq(a, b, h)
n <- length(x)

# Pre-allocate blank vector y
y <- numeric(n)

for (i in 1:n) {
  y[i] <- 0.85* x[i]^3 - 2 * x[i]^2 + rnorm(1, mean = 0, sd = 10.5)
}

plot(x, y, pch = 1)
```

```r
#Using ksmooth function, kernel = normal (Gaussian)

kernelreg_fit <- ksmooth(x, y,
                   kernel = "normal",
                   bandwidth = 1)

plot(x, y, pch = 19, col = "gray")
lines(kernelreg_fit$x, kernelreg_fit$y, col = "blue", lwd = 2)
```

```r
#Generating data for plotting
x_forpred <- seq(min(x), max(x), length = 150)

#Plotting
plot(x, y, pch = 19, col = "gray")

bw <- c(0.5, 1, 2)
cols <- c("pink", "light blue", "light green")

#Iterating for different bandwidth and then plotting
for (i in seq_along(bw)) {
  kernelreg_fit <-ksmooth(x, y,
                   kernel = "normal",
                   bandwidth = bw[i],
                   x.points = x_forpred)

  lines(kernelreg_fit$x, kernelreg_fit$y, col = colors[i], lwd = 3)
}

legend("topleft",
       legend = paste("bandwidth =", bw),
       col = cols, lwd = 3)
```

**Example 4: Regression Tree**

Here in this example, we will use the same data as in Example 2.

```r
# Generate sample data
a <- 0
b <- 6
h <- 0.05

x <- seq(a, b, h)
n <- length(x)

# Pre-allocate blank vector y
y <- numeric(n)

for (i in 1:n) {
  y[i] <- 0.85* x[i]^3 - 2 * x[i]^2 + rnorm(1, mean = 0, sd = 10.5)
}

dataxy<-data.frame(y, x)

plot(x, y, pch = 1)
```

Before we run the regression tree, we will be needing the library $\texttt{rpart}$ and $\texttt{rpart.plot}$

```r
install.packages("rpart")
install.packages("rpart.plot")
library("rpart")
library("rpart.plot")
```

Fitting our data to regression tree model

```r
#Fitting to regression tree model
treereg_fit<-rpart(y ~ x,
                   data = dataxy,
                   control = rpart.control(
                             cp = 0.01,
                             minsplit = 3)
)
```

```r
#Plotting the tree
rpart.plot(treereg_fit)
```

```r
#Plotting the predicted y values
new_xdata <- data.frame(
  x = seq(min(x) , max(x) , length = 150)
)

y_pred <- predict(treereg_fit, newdata = new_xdata)

#plotting points
plot(x, y, pch = 19, col = "gray")

#plotting the prediction line
lines(new_xdata$x, y_pred, col = "green", lwd = 3)
```

```r
printcp(treereg_fit)
plotcp(treereg_fit)
```

We focus on $\texttt{xerror}$  the cross - validation error and check which cp has a minimum xerror. In our case cp = 0.01 yields the minimum cross validation error
